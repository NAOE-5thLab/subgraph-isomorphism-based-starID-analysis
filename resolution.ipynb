```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from utils.font import font_setting

# os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
font_setting()


def make_patch_spines_invisible(ax):
    ax.set_frame_on(True)
    ax.patch.set_visible(False)
    for sp in ax.spines.values():
        sp.set_visible(False)
```

```python
def theta_max(theta_FOV, U):
    return np.arctan2(2*np.tan(theta_FOV/2), U)

theta_FOVs = np.linspace(0, np.deg2rad(100), 1000)

fig = plt.figure(figsize=(4.8, 3), tight_layout=True)
ax1 = fig.add_subplot(1,1,1)
kwargs = {"color": "black", "ls": "solid", "lw": 1.0, "label": "$U=1024$"}
ax1.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 1024)), **kwargs)
kwargs = {"color": "black", "ls": "dashed", "lw": 1.0, "label": "$U=2048$"}
ax1.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 2048)), **kwargs)
kwargs = {"color": "black", "ls": "dotted", "lw": 1.0, "label": "$U=4096$"}
ax1.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 4096)), **kwargs)
ax1.set_xlabel("$\\theta_{\\mathrm{FOV}} \\ \\mathrm{(degree)}$")
ax1.set_ylabel("$\\theta_{\\mathrm{res}} \\ \\mathrm{(degree)}$")
ax1.legend().set_alpha(1.0)

# ax2 = ax1.twinx()
# kwargs = {"color": "black", "ls": "-", "lw": 1.0, "label": "$U=1024$"}
# ax2.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 1024))*60, **kwargs)
# kwargs = {"color": "black", "ls": "--", "lw": 1.0, "label": "$U=2048$"}
# ax2.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 2048))*60, **kwargs)
# ax2.set_xlabel("$\\theta_{\\mathrm{FOV}} \\ \\mathrm{(degree)}$")
# ax2.set_ylabel("Posiotion error$\\ \\mathrm{(nm)}$")
# ax2.grid(visible=None)

# ax2.spines["left"].set_position(("axes", -0.25))
# make_patch_spines_invisible(ax2)
# ax2.spines["left"].set_visible(True)
# ax2.yaxis.set_label_position('left')
# ax2.yaxis.set_ticks_position('left')

ax3 = ax1.twinx()
kwargs = {"color": "black", "ls": "solid", "lw": 1.0, "label": "$U=1024$"}
ax3.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 1024))*60*1852, **kwargs)
kwargs = {"color": "black", "ls": "dashed", "lw": 1.0, "label": "$U=2048$"}
ax3.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 2048))*60*1852, **kwargs)
kwargs = {"color": "black", "ls": "dotted", "lw": 1.0, "label": "$U=4096$"}
ax3.plot(np.rad2deg(theta_FOVs), np.rad2deg(theta_max(theta_FOVs, 4096))*60*1852, **kwargs)
ax3.set_xlabel("$\\theta_{\\mathrm{FOV}} \\ \\mathrm{(degree)}$")
ax3.set_ylabel("Posiotion error$\\ \\mathrm{(m)}$")
ax3.grid(visible=None)

# ax3.spines["left"].set_position(("axes", -0.25)) 
# make_patch_spines_invisible(ax3)
# ax3.spines["left"].set_visible(True)
# ax3.yaxis.set_label_position('left')
# ax3.yaxis.set_ticks_position('left')

path = "./log/resolutions.pdf"
fig.savefig(path)
```
